```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1=cv2.imread('/content/download.jfif')
image=cv2.cvtColor (image1,cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
lap=cv2.Laplacian(image,cv2.CV_64F)
lap=np.uint8(np.absolute(lap))
plt.imshow(lap)
```

```python
im=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
plt.imshow(im)
```

```python
sobelx=cv2.Sobel(image1,cv2.CV_64F,1,0)
sobely=cv2.Sobel(image1,cv2.CV_64F,0,1)
sobelx=np.uint8(np.absolute(sobelx))
sobely=np.uint8(np.absolute(sobely))
sobelcombinbed=cv2.bitwise_or(sobelx,sobely)
plt.imshow(sobelx)
plt.imshow(sobely)
```

```python
img=cv2.imread('/content/download.jfif')
edge=cv2.Canny(img,100,200)
plt.imshow(img)
plt.title('original')
plt.show()
```

```python
plt.imshow(edge)
plt.title('canny edge dectection')
plt.show()
```
